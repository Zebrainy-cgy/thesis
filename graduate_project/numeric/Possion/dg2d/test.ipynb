```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
vertices = np.array([[0, 0, 0],
                     [1, 0, 0],
                     [0, 1, 0]])

# 
faces = [[vertices[j] for j in [0, 1, 2]]]

# 
triangle = Poly3DCollection(faces, facecolors='white', edgecolors='black')
ax.add_collection3d(triangle)

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('White Triangle Plane')

# 
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)

# 
plt.show()
```
