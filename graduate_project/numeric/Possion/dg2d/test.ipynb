```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# 创建一个新的图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 定义三角形的顶点
vertices = np.array([[0, 0, 0],
                     [1, 0, 0],
                     [0, 1, 0]])

# 定义三角形的面
faces = [[vertices[j] for j in [0, 1, 2]]]

# 绘制三角形
triangle = Poly3DCollection(faces, facecolors='white', edgecolors='black')
ax.add_collection3d(triangle)

# 设置坐标轴标签和标题
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('White Triangle Plane')

# 设置坐标轴范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)

# 显示图形
plt.show()
```
